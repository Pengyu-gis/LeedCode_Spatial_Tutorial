```python
import libpysal
```

```python
_ = libpysal.examples.load_example('South')
pth = libpysal.examples.get_path('south.shp')
```

```python
import geopandas
from geopandas import read_file
df = read_file(pth)
df
```

```python
%matplotlib inline
ax = df.plot(column='HR60')
```

```python
ax = df.plot(column='HR60', scheme='QUANTILES')
```

# **scheme: *str (default None)***


> Name of a choropleth classification scheme (requires mapclassify).



A mapclassify.MapClassifier object will be used under the hood. <br>

Supported are all schemes provided by mapclassify. <br>
(e.g. *‘BoxPlot’, ‘EqualInterval’, ‘FisherJenks’, ‘FisherJenksSampled’, ‘HeadTailBreaks’, ‘JenksCaspall’, ‘JenksCaspallForced’, ‘JenksCaspallSampled’, ‘MaxP’, ‘MaximumBreaks’, ‘NaturalBreaks’, ‘Quantiles’, ‘Percentiles’, ‘StdMean’, ‘UserDefined’*). <br>

 Arguments can be passed in classification_kwds.

```python
ax = df.plot(column='HR60', scheme='QUANTILES', k=4, cmap='BuPu')
```

# Add Legend

```python
ax = df.plot(column='HR60', scheme='QUANTILES', k=4, \
             cmap='BuPu', legend=True,
             legend_kwds={'loc': 'center left', 'bbox_to_anchor':(1,0.5),  'fmt':"{:.0f}"})
```

***legend_kwds: dict (default None)***<br>
Keyword argument to pass to **matplotlib.pyplot.legend()** or **matplotlib.pyplot.colorbar()**.
<br>

Additional accepted keywords when scheme is specified:<br>



1.   ***fmt:string***<br>
A formatting specification for the bin edges of the classes in the legend. For example, to have no decimals: {"fmt": "{:.0f}"}.
2.   ***labels:list-like***<br>
A list of legend labels to override the auto-generated labels. Needs to have the same number of elements as the number of classes (k).

3. ***interval:boolean (default False)***<br>
An option to control brackets from mapclassify legend. If True, open/closed interval brackets are shown in the legend.

```python
ax = df.plot(column='HR60', scheme='BoxPlot', \
             cmap='BuPu', legend=True,
             legend_kwds={'loc': 'center left', 'bbox_to_anchor':(1,0.5),
                         'fmt': "{:.0f}"})
```

```python
import mapclassify
bp = mapclassify.BoxPlot(df.HR60)
bp
```

```python
# Show Interval bracket

ax = df.plot(column='HR60', scheme='BoxPlot', \
             cmap='BuPu', legend=True,
             legend_kwds={'loc': 'center left', 'bbox_to_anchor':(1,0.5),
                         'interval': True})
```

categories: list-like<br>
Ordered list-like object of categories to be used for categorical plot.

```python
from pandas.core.arrays import categorical
ax = df.plot(column='STATE_NAME', categorical=True, legend=True, 
             legend_kwds={'loc':'center left', 'bbox_to_anchor':(1,0.5),
                         'fmt': "{:.0f}"})
```
